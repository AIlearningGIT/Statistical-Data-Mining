## ANSWER 2)

```r
ADVT<-data$ADVT*1000
```

```r
plot(ADVT,SALES, col = 'red')
abline(lm(SALES~ADVT), col = 'blue',lwd= 3, lty = 2)
```

```r
cor(ADVT,SALES)
```

## We are seeing a positive linear relationship between advertisement and sales but not that effective correlation between the two.
### 1) While some points tell us that advertisement has led to increase in sales in few districts, there are many districts where more money has been spent on advertisement and yet have not resulted in the growth in sales.
### 2) Some points where advertisement money spent is less than 8500 dollars, Sales are greater than 120000 dollars and in some districts where money spent on advertisement is greater than 11000 and yet sales have been less than 110000. So money spent on advertisements in those districts should be reduced.
### 3) There are exactly 4 districts where money spent on advertisements and sales are linearly correlated, in order to increase sales in those districts, there should be more expenditure on advertisements.

## ANSWER 3)

```r
library("lattice")
```

```r
INCOME_HIGH<-subset(data,INCOME>55.0)
```

```r
INCOME_LOW<-subset(data, INCOME<=55.0)
```

```r
par(mfrow=c(1,2))
xyplot(INCOME_HIGH$SALES*1000~INCOME_HIGH$ADVT*1000, main = 'HIGH INCOME ADVERTISEMENTS VS SALES', ylab = 'SALES', xlab = 'ADVT')
xyplot(INCOME_LOW$SALES*1000~INCOME_LOW$ADVT*1000, main = 'LOW INCOME ADVERTISEMENTS VS SALES',  ylab = 'SALES', xlab = 'ADVT')
```

```r
cor(INCOME_HIGH$SALES, INCOME_HIGH$ADVT)
```

```r
cor(INCOME_LOW$SALES, INCOME_LOW$ADVT)
```

# Three keypoints:
## 1) There is significantly high positive correlation between sales and advertisement money, and we can say that more investment on advertisement in these areas result to higher sales.
## 2) There is very loose positive correlation between sales and advertisement money, and we cannot say anything concrete about the relationship and instead we have to go more granular into the data before reaching to a conclusion.
## 3) The maximum sales of districts with higher income is close to 16000 dollars which is much highers than maximum sales of districts with low income and that we can clearly say that not only advertisements, but income also plays a crucial role in the sales of a product.
