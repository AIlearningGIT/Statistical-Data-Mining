# ANSWER 2)

```r
ROA_Bankrupt<-subset(data, Status == 'Bankrupt')
```

```r
ROA_Solvent<-subset(data, Status == 'Solvent')
```

```r
par(mfrow = c(1,2))
hist(ROA_Bankrupt$ROA)
hist(ROA_Solvent$ROA)
```

```r
cat(mean(ROA_Bankrupt$ROA),', ', mean(ROA_Solvent$ROA))
```

```r
t.test(ROA_Bankrupt$ROA)
```

```r
t.test(ROA_Solvent$ROA)
```

## Yes, both are statistically way different, mean ROA of bankrupt firms turn out to be '-0.0814' and solvent firms turn out to '+0.056'. This is indicative of the fact that firm which is bankrupt or about to go bankrupt can expect negative ROA's between -0.147 to -0.0154 with 95% confidence interval.
